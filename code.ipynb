```python
import numpy as np
import pandas as pd
import ast
```

```python
df_movie=pd.read_csv("D:/Data_science/Movie Recomdation System/tmdb_5000_movies.csv/tmdb_5000_movies.csv")
df_credit=pd.read_csv("D:/Data_science/Movie Recomdation System/tmdb_5000_credits.csv/tmdb_5000_credits.csv")
```

```python
df_credit.info()
```

```python
movie=df_movie.merge(df_credit,on='title')
movie.info()
```

```python
movie=movie[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movie.dropna(inplace=True)
```

```python
movie['cast']
```

```python
def modify(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movie['genres']=movie['genres'].apply(modify)
```

```python
movie['keywords']=movie['keywords'].apply(modify)
```

```python
def convert(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter<3:
            L.append(i['name'])
            counter=counter+1
    return L 
```

```python
movie['cast']=movie['cast'].apply(convert)
```

```python
movie['crew'][0]
```

```python
def director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L            
```

```python
movie['crew']=movie['crew'].apply(director)
```

```python
def space_remove(text):
    L=[]
    for i in text:
        L.append(i.replace(" ",""))
    return L
```

```python
movie['cast'] = movie['cast'].apply(space_remove)
movie['crew'] = movie['crew'].apply(space_remove)
movie['genres'] = movie['genres'].apply(space_remove)
movie['keywords'] = movie['keywords'].apply(space_remove)
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie['overview']
```

```python
movie['tag']=movie['overview']+movie['genres']+movie['keywords']+movie['cast']+movie['crew']
```

```python
new = movie.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tag'] = new['tag'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tag']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recomdetion(movie):
    index=new[new['title']==movie].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distance[1:6]: # top 5 movie only
        print(new.iloc[i[0]].title)
    
        
    
    
```
